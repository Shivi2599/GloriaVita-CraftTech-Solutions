```python
pip install playsound2

from tkinter import *
from tkinter import messagebox 
import time
import threading
from playsound import playsound

# Create a window
window = Tk()
window.title('Countdown Timer')  # Title of Window
window.config(bg="#2A2C2F")      # Background Color
window.geometry("430x600")       # Size of Window
window.resizable(False, False)   # Prohibit resizing of the window

# Validation function to allow only 2 characters
def validate_input(new_value):
    if new_value == "":  # Allow empty input
        return True
    return new_value.isdigit() and 0 <= int(new_value) <= 59 and len(new_value) <= 2

validate_cmd = window.register(validate_input)

# Heading 
Heading = Label(window, text='Timer', font="calibri 30 bold", fg='#DB0F27', bg='#2A2C2F')
Heading.pack(pady=10)

# Current Time
current_time_label = Label(window, text='Current Time:', font="calibri 18 bold", fg='#eeeeee', bg='#2A2C2F')
current_time_label.place(x=80, y=90)

def digital_clock():
    current_time = time.strftime("%H:%M:%S")
    clock_time.config(text=current_time)
    clock_time.after(1000, digital_clock)

clock_time = Label(window, text="", font="calibri 18 bold", fg='#eeeeee', bg='#2A2C2F')
clock_time.place(x=230, y=90)
digital_clock()

# Timer Input Fields
hours = StringVar(value="00")
minutes = StringVar(value="00")
seconds = StringVar(value="00")

# Hour Entry
hours_entry = Entry(window, font="calibri 50", validate="key", validatecommand=(validate_cmd, "%P"), textvariable=hours, width=2, bg='#eeeeee', fg='black')
hours_entry.place(x=50, y=160)

# Label ":"
Label(window, text=":", font="calibri 60 bold", fg='#eeeeee', bg='#2A2C2F').place(x=140, y=160, height=75, width=20)

# Minute Entry
minutes_entry = Entry(window, font="calibri 50", validate="key", validatecommand=(validate_cmd, "%P"), textvariable=minutes, width=2, bg='#eeeeee', fg='black')
minutes_entry.place(x=180, y=160)

# Label ":"
Label(window, text=":", font="calibri 60 bold", fg='#eeeeee', bg='#2A2C2F').place(x=270, y=160, height=75, width=20)

# Second Entry
seconds_entry = Entry(window, font="calibri 50", validate="key", validatecommand=(validate_cmd, "%P"), textvariable=seconds, width=2, bg='#eeeeee', fg='black')
seconds_entry.place(x=310, y=160)

# Labels
Label(window, text='hrs', font="calibri 18", fg='#eeeeee', bg='#2A2C2F').place(x=70, y=250)
Label(window, text='mins', font="calibri 18", fg='#eeeeee', bg='#2A2C2F').place(x=188, y=250)
Label(window, text='secs', font="calibri 18", fg='#eeeeee', bg='#2A2C2F').place(x=325, y=250)

# Global variables
paused = False  
times = 0  

# Start function
def start():
    global times, paused
    try:
        h = int(hours.get() or 0)
        m = int(minutes.get() or 0)
        s = int(seconds.get() or 0)
        times = h * 3600 + m * 60 + s  

        if times <= 0:
            messagebox.showerror("Error", "Enter a valid time!")
            return

        paused = False
        countdown()
    except ValueError:
        messagebox.showerror("Error", "Please enter valid numbers.")

# Countdown function
def countdown():
    global times, paused
    if times >= 0 and not paused:
        h, rem = divmod(times, 3600)
        m, s = divmod(rem, 60)

        hours.set(f"{h:02d}")
        minutes.set(f"{m:02d}")
        seconds.set(f"{s:02d}")

        times -= 1
        window.after(1000, countdown)
    elif times < 0:
        messagebox.showinfo("Time's Up!", "Countdown Finished!")
        threading.Thread(target=playsound, args=(r"D:\SHIVANGI\Interns\GloriaVita CraftTech Solutions\Countdown Timer\alarm sound effects.wav",), daemon=True).start()
        reset()

# Pause function
def pause():
    global paused
    paused = not paused  
    pause_button.config(text="Resume" if paused else "Pause")
    if not paused:
        countdown()

# Reset function
def reset():
    global times, paused
    times = 0
    paused = False
    hours.set("00")
    minutes.set("00")
    seconds.set("00")
    pause_button.config(text="Pause")

# Buttons
start_button = Button(window, text='Start', font="calibri 20 bold", fg='black', bg='#eeeeee', command=start)
start_button.place(x=90, y=350, height=50, width=110)

pause_button = Button(window, text='Pause', font="calibri 20 bold", fg='black', bg='#eeeeee', command=pause)
pause_button.place(x=240, y=350, height=50, width=110)

reset_button = Button(window, text='Reset', font="calibri 20 bold", fg='black', bg='#eeeeee', command=reset)
reset_button.place(x=165, y=450, height=50, width=110)

# Run
window.mainloop()
```
